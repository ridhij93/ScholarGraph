```python
import requests
from bs4 import BeautifulSoup
import re
import matplotlib.pyplot as plt

def get_paper_count(topic, start_year, end_year):
    
#         base_url = f"https://scholar.google.com/en/scholar"  # Use the English version of Google Scholar
#     headers = {
#         "User-Agent": "Your User Agent Here",  # Replace with your user agent
#         "Accept-Language": "en-US,en;q=0.9",  # Set the preferred language to English
#     }
    
    base_url = f"https://scholar.google.com/scholar" 
    headers = {"User-Agent": "Your User Agent Here",
              "Accept-Language": "en-US,en;q=0.9",}  # Replace with your user agent
    
    years = list(range(start_year, end_year + 1))
    paper_counts = []

    for year in years:
#         &as_epq=&as_oq=&as_eq=&as_occt=any&as_sauthors=&as_publication=&as_ylo=2000&as_yhi=2000&hl=en&as_sdt=0%2C5
        query = f"as_q={topic}+&as_epq=&as_oq=&as_eq=&as_occt=any&as_sauthors=&as_publication=&as_ylo={year}&as_yhi={year}&hl=en&as_sdt=0%2C5"
        url = f"{base_url}?{query}"
        print (url)
        response = requests.get(url, headers=headers)
        response.encoding = 'utf-8'
        
        soup = BeautifulSoup(response.text, 'html.parser')
        print (soup)
        # Extract the approximate number of results from the search page
        result_stats = soup.find(id="gs_ab_md")
#         if result_stats == None:
#             result_stats = soup.find(id="gs_ab_mdw")
        print (result_stats)
        if result_stats:
            count_text = result_stats.text
            count_text = count_text.replace(",", "")
            
            print ("count ", count_text)
            # Use regular expressions to extract numeric values
            numeric_values = re.findall(r'\d+', count_text)
            print ("Num ", numeric_values[0])
            
            if len(numeric_values) >= 1:
                count = int(numeric_values[0])
            else:
                count = 0
        else:
            count = 0
        
        paper_counts.append(count)
    print (years , " -- ", paper_counts)
    return years, paper_counts



def plot_paper_counts(years, counts, topic):
    plt.figure(figsize=(10, 6))
    plt.bar(years, counts, color='skyblue')
    plt.xlabel("Year")
    plt.ylabel("Number of Papers")
    plt.title(f"Number of Papers on '{topic}' Published Each Year")
    plt.xticks(years)
    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    topic = "large+language+model"
    start_year = 2000
    end_year = 2022

    years, paper_counts = get_paper_count(topic, start_year, end_year)
    plot_paper_counts(years, paper_counts, topic)
```
